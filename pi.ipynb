```
import random
import pandas as pd

from pyspark.sql.types import *
from pyspark.sql.functions import col,udf
import time
from pyspark.sql.functions import pandas_udf, PandasUDFType                                                                                                                                           


start_time = time.time()

@pandas_udf('float', PandasUDFType.SCALAR)
def surface_area_pandas2(height, radian):
  pi=22/7
  volume = pi * radian * radian * height
  sur_area = ((2*pi*radian) * height) + ((pi*radian**2)*2)
  return sur_area
                                                                                                                                                                                                          
@pandas_udf('float', PandasUDFType.SCALAR)
def surface_area_pandas(height, radians):
  calc_df = pd.DataFrame({'height1' : height, 'radians1' : radians})
  calc_df['surface_area'] = calc_df.apply(lambda x : calc_pi(x['height1'], x['radians1']), axis=1)
  return calc_df['surface_area']
                 
def surface_area(height, radians):
  return calc_pi(height, radians)


def calc_pi(height, radians):
  pi=22/7
  height = float(height)
  radian = float(radians)
  volume = pi * radian * radian * height
  sur_area = ((2*pi*radian) * height) + ((pi*radian**2)*2)
  return sur_area


mySchema = StructType((StructField("height", LongType(), True),StructField("radians", LongType(), True)))
tuplelist = [(random.randint(x, x+1000),random.randint(x, x+200)) for x in range(1, 10000000)]
valsRDD = spark.sparkContext.parallelize(tuplelist,500)
type(valsRDD)
valsDF = spark.createDataFrame(valsRDD, mySchema)
#newDF = valsDF.withColumn("surface_area", surface_area_pandas2(col("height"), col("radians")))
myUDF = udf(surface_area)
newDF = valsDF.withColumn("surface_area", myUDF(col("height"), col("radians")))
display(newDF)

print("--- %s seconds ---" % (time.time() - start_time))
```
